#### a) Using the Sales Target dataset, calculate the percentage change in target sales for the Furniture category month-over-month. 

```python
import pandas as pd
```

```python
sales_target_df = pd.read_csv("Sales_target_1C8295CCDE.csv")
```

```python
sales_target_df.head()
```

```python
sales_target_df.shape
```

```python
sales_target_df["percentage_change"] = sales_target_df["Target"].pct_change() * 100
```

```python
sales_target_df["percentage_change"].fillna(0,inplace=True)
sales_target_df
```

#### b) Analyse the trends to identify months with significant target fluctuations. Suggest strategies for aligning target expectations with actual performance trends.

```python
threshold = 0.10
significant_fluctuations = sales_target_df[sales_target_df["percentage_change"].abs() > threshold]
significant_fluctuations
```

#### What does above data tell?
#### For Clothing July-18 & October-18 showed the highest percentage change while April-18 showed the lowest
#### For Electronics There is huge fluctuation between Jan-19 & Apr-18
